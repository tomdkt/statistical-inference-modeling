```python
#@title Bibliotecas
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf  #adicionado na semana 6
```

```python
lstx1 = (2.04, 2.04, 3.06, 3.06, 4.08, 4.08, 2.06, 2.06,
         3.08, 3.08, 4.11, 4.11, 2.01, 3.02, 4.03)
lstx2 = (3.55, 6.07, 3.55, 6.07, 3.55, 6.16, 3.62, 6.16,
         3.62, 5.89, 3.62, 5.89, 6.18, 6.18, 6.18)
lstx3 = np.multiply(lstx1, lstx2)
lsty = (3.11, 3.26, 3.89, 10.25, 3.11, 13.48, 3.94, 3.53,
        3.36, 6.49, 2.72, 12.48, 2.82, 5.19, 8.04)
# Construir o DataFrame e nomear as colunas
df = pd.DataFrame(list(zip(lstx1, lstx2, lstx3, lsty)),
                  columns=["x1", "x2", "x3", "y"])
```

```python
regmul = smf.ols('y ~ x1 + x2 + x3', data=df)  #members of reg object print(dir(reg))
res = regmul.fit()
```

```python
#@title Resultado detalhado
print(res.summary())
```

```python
#@title Estatística de teste f
#Estatística f calculada no slide 14 
print(res.fvalue)  # este valor deve ser menor do que o Fcrit para aceitar H0
print(res.f_pvalue)  # este valor deve ser maior do que o alpha para aceitar H0
```

```python
#@title complementares
#ANOVA
table = sm.stats.anova_lm(res, typ=1)  #
print(table)
```

```python
#ANOVA
table = sm.stats.anova_lm(res, typ=1)  # typ pode ser 2 ou 3
totalSM = np.sum(table.iloc[0:3, 1])
totalMean = np.sum(table.iloc[0:3, 1]) / np.sum(table.iloc[0:3, 0])
print(table.iloc[3, 0:4])
print("SQE=", totalSM)
print(totalMean)
```

```python
#Se precisar calcular IC de algum parâmetro
k = res.df_model  # grau do modelo
n = res.nobs  # num. amostras
#usar a tabela tstudent para t
from scipy.stats import t

alpha = 0.01  # significancia = 99%
df = n - (k + 1)  # graus de liberdade
v = t.ppf(1 - alpha / 2, df)
tt = v
print(f't_crit=: {v}')
```

```python
lstx1 = (2.04, 2.04, 3.06, 3.06, 4.08, 4.08, 2.06, 2.06,
         3.08, 3.08, 4.11, 4.11, 2.01, 3.02, 4.03)
lstx2 = (3.55, 6.07, 3.55, 6.07, 3.55, 6.16, 3.62, 6.16,
         3.62, 5.89, 3.62, 5.89, 6.18, 6.18, 6.18)
lstx3 = np.multiply(lstx1, lstx1)
lstx4 = np.multiply(lstx2, lstx2)
lstx5 = np.multiply(lstx1, lstx2)
lsty = (3.11, 3.26, 3.89, 10.25, 3.11, 13.48, 3.94, 3.53,
        3.36, 6.49, 2.72, 12.48, 2.82, 5.19, 8.04)
# Construir o DataFrame e nomear as colunas
df = pd.DataFrame(list(zip(lstx1, lstx2, lstx3, lstx4, lstx5, lsty)),
                  columns=["x1", "x2", "x3", "x4", "x5", "y"])
```

```python
regmul = smf.ols('y ~ x1 + x2 + x3 + x4 + x5', data=df)
res = regmul.fit()
print(res.summary())
```

```python
k = res.df_model  # grau do modelo
n = res.nobs  # num. amostras
#usar a tabela tstudent para t
from scipy.stats import t

alpha = 0.05  # significia = 5%
df = n - (k + 1)  # graus de liberdade
v = t.ppf(1 - alpha / 2, df)
tt = v
print(f't_crit=: {v}')
```
